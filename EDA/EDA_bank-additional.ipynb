```python
cols_numericas = ['cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
```

```python
for col in cols_numericas: #creo un bucle for para cambiar las , por .  ya que en el resultado de la tabla me salía NA esas columnas al hacer el cambio de tipo de dato
        df_bank[col] = df_bank[col].str.replace(',', '.').str.strip()
```

```python
df_bank
```

```python
df_bank[cols_numericas] = df_bank[cols_numericas].apply(pd.to_numeric, errors='coerce') #cambio de tipo de dato de object a numero en columnas numericas
df_bank
```

```python
df_bank.info()
```

```python
# Reemplazar los meses en español por inglés, ya que si no, no se puede hacer el camvio de dato a la columna
meses = {
    'enero': 'January', 'febrero': 'February', 'marzo': 'March',
    'abril': 'April', 'mayo': 'May', 'junio': 'June',
    'julio': 'July', 'agosto': 'August', 'septiembre': 'September',
    'octubre': 'October', 'noviembre': 'November', 'diciembre': 'December'
}

df_bank['date'] = df_bank['date'].str.lower()  # Por si hay mayúsculas
for es, en in meses.items():
    df_bank['date'] = df_bank['date'].str.replace(es, en, regex=False)
```

```python
df_bank['date'] = pd.to_datetime(df_bank['date']) #cambiamos la columna de fecha
df_bank
```

```python
cols_cat = ['job', 'marital', 'education', 'contact', 'poutcome']
df_bank[cols_cat] = df_bank[cols_cat].astype('category')
```

```python
df_bank['age'] = df_bank['age'].astype('Int64') 
```
